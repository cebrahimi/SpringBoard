```python
species = df.SpeciesIDDesc
species = species.dropna() #drop nan values in species feature
speciesOfAnimal = species.unique()
print(speciesOfAnimal)
```

```python
animal_list = []
for  i in speciesOfAnimal:
    animal_list.append(len(species[species==i]))
ax = sns.barplot(x=speciesOfAnimal, y =animal_list)
plt.title('Number of Species Bite')
plt.xticks(rotation=90)
print(animal_list)
```

```python
def animal_month(animal,data):
    month_list= ['01','02','03','04','05','06','07','08','09','10','11','12']
    numberOfAnimal = []
    for i in month_list:
        x = df.loc[(df['SpeciesIDDesc']==animal)&(df['bite_date'].str.split('-').str[1]==i)]
        numberOfAnimal.append(len(x))
    ax = sns.barplot(x=month_list,y=numberOfAnimal)
    plt.title(animal + ' bite for 12 month')
```

```python
animal_month('DOG',df)
```

```python
animal_month('CAT',df)
```

```python
animal_month('BAT',df)
```

```python
dogType = df[df['SpeciesIDDesc'] == 'DOG']
dogTypecount = dogType['BreedIDDesc'].value_counts()
dogTypecount
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x = dogTypecount.index, y = dogTypecount)
plt.title("Dog breeds involved in bites", fontsize=20)
plt.xlabel('Dog breeds', fontsize=18)
plt.ylabel('Bites', fontsize=18)
plt.xticks(fontsize=10, rotation=90)
plt.show()
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x = dogTypecount[:15].index, y = dogTypecount[:15])
plt.title("Top 15 dogs involved in bites", fontsize=20)
plt.xlabel('Dog breeds', fontsize=18)
plt.ylabel('Bites', fontsize=18)
plt.xticks(fontsize=10, rotation=90)
plt.show()
```

```python
dogGendercount = dogType['GenderIDDesc'].value_counts()
dogGendercount
```

```python
#plt.figure(figsize = (20,10))
sns.barplot(x = dogGendercount.index, y = dogGendercount)
plt.title("Dog gender involved in bites", fontsize=20)
plt.xlabel('Dog genders', fontsize=18)
plt.ylabel('Bites', fontsize=18)
plt.xticks(fontsize=10, rotation=90)
plt.show()
```
