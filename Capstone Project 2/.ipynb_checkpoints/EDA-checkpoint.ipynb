```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import json
```

```python
df = pd.read_csv('cleanDF.csv', index_col = 0, parse_dates = ['publish_time'])
```

```python
df.head()
```

```python
df.Category.value_counts()
```

```python
df['Category'].value_counts().head(15).sort_values().plot(kind='barh', figsize=(20,15))
plt.title("Top 15 Youtube Categories - the most frequrently appearing in Trending Videos", size=20)
plt.show()
```

```python
df['publish_month'] = df['publish_time'].dt.month
```

```python
f, ax = plt.subplots(4,1,figsize=(20,15))
g = sns.barplot(x=df['publish_month'],y=df['views'],data=df,ax=ax[0])
g.set_title('Views across published months')
g.set_xlabel('')
g1 = sns.barplot(x=df['publish_month'],y=df['comment_count'],data=df,ax=ax[1])
g1.set_title('Comments across published months')
g1.set_xlabel('')
g2 = sns.barplot(x=df['publish_month'],y=df['likes'],data=df,ax=ax[2])
g2.set_title('Likes across published months')
g2.set_xlabel('')
g3 = sns.barplot(x=df['publish_month'],y=df['dislikes'],data=df,ax=ax[3])
g3.set_title('Dislikes across published months')
plt.show()
```

It seems to appear that the summer month attract more visits on youtube home page, with most amount of views, likes, dislikes, and comments.

```python
f, ax = plt.subplots(4, 1, figsize=(22, 28))

sns.barplot(y=df['Category'],x=df['views'],data=df,ax=ax[0])
sns.barplot(y=df['Category'],x=df['comment_count'],data=df,ax=ax[1])
sns.barplot(y=df['Category'],x=df['likes'],data=df,ax=ax[2])
sns.barplot(y=df['Category'],x=df['dislikes'],data=df,ax=ax[3])
```

-Category x Views: It seems that music has the most amount of views, probably due to people watching music videos on release to see the choreography. Following music is Film & Animation, Nonprofit & Activism, and gaming. These four seem to be the most popular and typical video types that would appear on the trending youtube home-page

-Category x Comment Count: People tend to comment more on Nonprofit and Activism type of video to show support of the video.

-Category x Likes: Nonprofit & Activism and Music were the two most popular categories in terms of likes recieved on the video.

-Category x dislikes: Nonprofit & Activism and gaming were the two categories where it contained the most amount of dislikes as well.

Overall it seems that Nonprofit & Activism has the most fluctuation when it comes to numerical analysis of youtube trending vidoes stats (views, amount of comments, likes, dislikes).

```python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(df.corr(), annot=True, linewidths=.5, ax=ax)
```

There is a strong correlation between the amount of views and likes, as well as likes and comment count.

A somewhat strong correlation between views and comment count, and dislikes and comment count

```python
f, ax = plt.subplots(2,2,figsize=(20,15))
ax[0][0] = sns.scatterplot(x=df['likes'],y=df['views'],data=df,ax=ax[0][0])
ax[0][0].set_title('Correlation of Likes and Views')

ax[0][1] = sns.scatterplot(x=df['likes'],y=df['comment_count'],data=df,ax=ax[0][1])
ax[0][1].set_title('Correlation of Likes and Comment Counts')

ax[1][0] = sns.scatterplot(x=df['dislikes'],y=df['comment_count'],data=df,ax=ax[1][0])
ax[1][0].set_title('Correlation of Dislikes and Comment Counts')

ax[1][1] = sns.scatterplot(x=df['likes'],y=df['dislikes'],data=df,ax=ax[1][1])
ax[1][1].set_title('Correlation of Likes and Dislikes')
```

```python
df["publish_hour"] = df["publish_time"].apply(lambda x: x.strftime('%H'))

hour_summary = df["publish_hour"].value_counts().to_frame().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
ax = sns.barplot(x="index", y="publish_hour", data=hour_summary, ax=ax)
_ = ax.set(xlabel="Hour", ylabel="No. of videos")
_ = ax.set_title('Number of videos by hour', fontsize =16)
_ = plt.show()
```

The hours of 3pm to 6pm seems to be the most likely time where youtube vidoes are posted, that became trending.
